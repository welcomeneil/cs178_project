```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

```

```python
# loading data and train set splitting
fashion_mnist = tf.keras.datasets.fashion_mnist

(train_X, train_y), (test_X, test_y) = fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

```

```python
opt = ['adam','sgd','RMSprop','nadam', 'adamax','adagrad']
train_scale = [train_X, train_X / 255.0]
test_scale = [test_X, test_X / 255.0]
num_epoch = [5, 10, 15 ,20, 25]
train_accus = [[],[]]
test_accus = [[],[]]
for i in range(len(opt)):
    for k in range(2):
        train_accu = []
        test_accu = []
        for n in num_epoch:
            model = tf.keras.Sequential([
            tf.keras.layers.Flatten(input_shape=(28, 28)),
            tf.keras.layers.Dense(128, activation='relu'),
            tf.keras.layers.Dense(10)
            ])
            model.compile(optimizer=opt[i],
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
            model.fit(train_scale[k], train_y, epochs=n)
            _, train_acc = model.evaluate(train_scale[k], train_y, verbose=2)
            train_accu.append(train_acc)
            _, test_acc = model.evaluate(test_scale[k], test_y, verbose=2)
            test_accu.append(test_acc)
        train_accus[k].append(train_accu)
        test_accus[k].append(test_accu)


```

```python
color = ['y','r','blue','green','pink','purple']
figure, axes = plt.subplots() 
axes.set_title("Neural Network train accuracy")
for i in range(6):
    if i != 1:
        axes.set_xlabel("num of epochs")
        axes.set_ylabel("accuracy")
        axes.plot(num_epoch,train_accus[0][i],c=color[i],label=opt[i]+' + unscaled',linestyle="dashed")
        axes.plot(num_epoch,train_accus[1][i],c=color[i],label=opt[i]+' + scaled')
axes.legend(bbox_to_anchor=(1., 1.05),
          )
```

```python
figure, axes = plt.subplots() 
axes.set_title("Neural Network test accuracy")
for i in range(6):
    if i != 1:
        axes.set_xlabel("num of epochs")
        axes.set_ylabel("accuracy")
        axes.plot(num_epoch,test_accus[0][i],c=color[i],label=opt[i]+' + unscaled',linestyle="dashed")
        axes.plot(num_epoch,test_accus[1][i],c=color[i],label=opt[i]+' + scaled')
axes.legend(bbox_to_anchor=(1., 1.05),
          )
```

```python
figure, axes = plt.subplots() 
axes.set_title("SGD optimizor accuracy")
axes.set_xlabel("num of epochs")
axes.set_ylabel("accuracy")
axes.plot(num_epoch,test_accus[0][1],c='r',label='test + unscaled',linestyle="dashed")
axes.plot(num_epoch,test_accus[1][1],c='r',label= 'test + scaled')
axes.plot(num_epoch,train_accus[0][1],c='blue',label='train + unscaled',linestyle="dashed")
axes.plot(num_epoch,train_accus[1][1],c='blue',label= 'train + scaled')
axes.legend(bbox_to_anchor=(1., 1.05),
          )
```
